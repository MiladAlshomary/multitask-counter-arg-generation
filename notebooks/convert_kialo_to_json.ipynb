```python
claims_tree = list()
```

```python
import os
for filename in os.listdir("../kialo_corpus"):
    data = json.load(open("../kialo_corpus/"+filename))
    df = pd.json_normalize(data)
    df_locations = pd.DataFrame(df['arguments.locations'][0])
    df_claims = pd.DataFrame(df['arguments.claims'][0])
    
    # take out the list of claims to be processed.
    for parent_id, text in zip(df_claims['id'], df_claims['text']):

        # extract children of this claim
        child_ids = list(df_locations[df_locations['parentId']==parent_id].targetId.values)
        child_relations = list(df_locations[df_locations['parentId']==parent_id].relation.values)

        # add this parent claim and its children in a dictionary
        parent_obj = {'id' : parent_id, 'text': text, 'pro': list(), 'con': list(), 'neutral':list()}

        # iterate among the child objects and add them to the parent object one by one.
        for child_id, relation in zip(child_ids, child_relations):
            # extract text related to that child id.
            text = df_claims[df_claims['id']==child_id].text.values[0]
            child_obj = {'id': child_id, 'text':text}
            if int(relation) == 1:
                parent_obj['pro'].append(child_obj)
            elif int(relation) == -1:
                parent_obj['con'].append(child_obj)
            else:
                parent_obj['neutral'].append(child_obj)  # this case is only for debugging.


        claims_tree.append(parent_obj)



```

```python
import json
# output_file = open('/home/akshit/Documents/job/ExplainConcepts/json/Astrophysict.txt','w')
# a = json.dumps(main_dict)

with open('../kialo_processing_akshit/kialo.txt','w') as json_file:
    json.dump(claims_tree, json_file)
```
